```python
import os
from shutil import copy2
from collections import defaultdict

def trouver_et_afficher_doublons(repertoires_sources, repertoire_cible):
    fichiers_plus_recents = {}  
    chemins_doublons = defaultdict(list)  

    for repertoire_source in repertoires_sources:
        for racine, dossiers, fichiers in os.walk(repertoire_source):
            for fichier in fichiers:
                chemin_complet = os.path.join(racine, fichier)
                if not os.path.isfile(chemin_complet):
                    continue

                #on sépare le nom du fichier de son extension
                nom_base, _ = os.path.splitext(fichier)

                chemins_doublons[nom_base].append(chemin_complet)

                if nom_base in fichiers_plus_recents:
                    if os.path.getmtime(chemin_complet) > os.path.getmtime(fichiers_plus_recents[nom_base]):
                        fichiers_plus_recents[nom_base] = chemin_complet
                else:
                    fichiers_plus_recents[nom_base] = chemin_complet

    for nom_base, chemin_le_plus_recent in fichiers_plus_recents.items():
        doublons = [chemin for chemin in chemins_doublons[nom_base] if chemin != chemin_le_plus_recent]
        if doublons:
            print(f"Cas de doublons observé pour '{nom_base}':")
            print(f"Fichier le plus récent : {chemin_le_plus_recent}")
            print(f"Fichiers doublons moins récents : {' '.join(doublons)}")
            print("---")


        _, ext = os.path.splitext(chemin_le_plus_recent)
        chemin_cible_complet = os.path.join(repertoire_cible, nom_base + ext)
        print(f"Copie de {chemin_le_plus_recent} vers {chemin_cible_complet}")
        copy2(chemin_le_plus_recent, chemin_cible_complet)


repertoires_sources = [
    '/Users/mourad/Documents/test_marouane_2',
    '/Users/mourad/Documents/test_marouane_1',
    '/Users/mourad/Documents/test_marouane_3'
    
]
repertoire_cible = '/Users/mourad/Documents/test_marouane_cible'
trouver_et_afficher_doublons(repertoires_sources, repertoire_cible)

```
